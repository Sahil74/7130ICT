```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Circle
#Load job market
```

```python
%matplotlib inline
df = pd.read_csv('job-market.csv')
```

```python
df.head(25)
```

```python
sns.countplot(data=df, y="Location")
```

```python
#df['Salary_ranges'] = '{}-{}'.format(df['LowestSalary'],df['HighestSalary'])
a =[]
for i in range(0,len(df['Date'])):
    a.append('{}-{}'.format(df['LowestSalary'][i],df['HighestSalary'][i]))
df['Salary_ranges'] = a
```

```python
df['Percentage'] = (df['HighestSalary']-df['LowestSalary'])/100
```
